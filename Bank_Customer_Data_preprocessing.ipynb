#Above is the best example for business validation before outlier analysis

Looking at the above result,
invalid credit score ranges from 902 to 949

these are not extreme values like 1200 or 2000
First thing is to know the Business defination of  this score.

Step 1: Verify Dataset Documentation

Some datasets use:

Score System	Range:
FICO	300-850,
Generic Credit Score	300-900,
Internal Bank Score	300-1000,
Custom Risk Score	0-1000 .

If the dataset creator[Risk Team] allows scores up to 1000, then these 927–949 values are perfectly valid.We can check once relationship with default



```python
#Check relationship with default

df.groupby('credit_card_default')['credit_score'].describe()
```

#Here Higher scores are corresponds to lower default risk,then these features are behaving correctly

Customers who did not default have an average credit score of 797, while customers who defaulted have an average score of 615. This substantial difference indicates that credit score is a strong predictor of default risk and should be retained as an important feature in the model.

This suggests the scores above 900 are behaving logically:

Very high scores belong to low-risk customers. They are helping separate defaulters from non-defaulters.

Therefore, from a modeling perspective:

✅ Keep the credit_score feature as-is.

✅ Do not remove those records.

✅ Do not cap the scores unless business documentation explicitly says the score range must end at 900.

```python
outlier_summary = []

for col in continuous_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = len(df[(df[col] < lower) | (df[col] > upper)])

    outlier_summary.append([col, count])

outlier_df = pd.DataFrame(
    outlier_summary,
    columns=['Column', 'Outlier_Count']
)

print(outlier_df)
```

#Feature Engineering:
Creating new useful features from existing data to improve model performance.
Example in Bank Defaulter Dataset

```python
#1.Debit to income ratio

df['Debit_income_ratio']=(df['yearly_debt_payments']/df['net_yearly_income'])

df['Debit_income_ratio']
```

```python
#For our  bank defaulter prediction project, lets calculate the Debt-to-Income (DTI) ratio in % and then categorize customers into risk groups.

#Step 1: Calculate DTI %
df['dti_ratio %']=(df['yearly_debt_payments']/df['net_yearly_income'])*100
df[['customer_id','yearly_debt_payments','net_yearly_income','dti_ratio %']].head()

```

```python
#Step 2: Categorize Risk

df['dti_category']=pd.cut(df['dti_ratio %'],bins=[0,35,50,float('inf')],labels=['Good Risk','Medium Risk','high Risk'])
#pd.cut() divides a numerical column into groups (bins) and assigns a label to each group.

df['dti_category'].value_counts()
```

Good Risk (< 35%) → Customer uses a small portion of income for debt repayment.
Medium Risk (35%–50%) → Customer has a moderate debt burden.
High Risk (> 50%) → More than half of income goes toward debt; higher chance of default.

```python
#Analyze Default Rate by Risk Category
pd.crosstab(df['dti_category'],df['credit_card_default'])
```

```python
df['dti_ratio %'].describe()
```

```python
df['dti_ratio %'].hist(bins=20)
```

```python
#instead of fixed threshold
df['dti_category'] = pd.qcut(
    df['dti_ratio %'],
    q=3,
    labels=['Low','Medium','High']
)
```

```python
df['dti_category'].value_counts()
```

```python
pd.crosstab(
    df['dti_category'],
    df['credit_card_default'],
    normalize='index'
) * 100
```

```python
#credit utilisation category

df["credit_utilization_category"] = pd.cut(
   df["credit_limit_used(%)"],
   bins=[0,30,70,100],
   labels=["Low","Medium","High"]
)
```

```python

df["credit_utilization_category"].value_counts()
```

```python
pd.crosstab(
    df['credit_utilization_category'],
    df['credit_card_default']
)
```

```python
# Age Group category
df["age_group"] = pd.cut(
   df["age"],
   bins=[18,30,45,60,100],
   labels=["Young","Adult","Middle","Senior"]
)
```
